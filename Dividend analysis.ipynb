```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the ticker symbol
ticker = 'MSFT'

# Fetch the data
msft = yf.Ticker(ticker)

# Fetch dividend data
dividends = msft.dividends

# Filter the data for the period from 2015-01-01 to 2023-12-31
dividends = dividends[(dividends.index >= '2015-01-01') & (dividends.index <= '2023-12-31')]

# Resample dividends data to quarterly frequency and sum up dividends for each quarter
quarterly_dividends = dividends.resample('Q').sum()

# Plot dividends
plt.figure(figsize=(10, 6))
plt.plot(quarterly_dividends.index, quarterly_dividends, marker='o', linestyle='-', color='b')
plt.title('Microsoft Quarterly Dividends (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Dividends (USD)')
plt.grid(True)
plt.xticks(quarterly_dividends.index, [date.strftime('%Y-%m') for date in quarterly_dividends.index], rotation=45)
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the ticker symbols
tickers = ['AAPL', 'MSFT']

# Fetch the data for both tickers
aapl = yf.Ticker(tickers[0])
msft = yf.Ticker(tickers[1])

# Fetch dividend data
aapl_dividends = aapl.dividends
msft_dividends = msft.dividends

# Filter the data for the period from 2015-01-01 to 2023-12-31
aapl_dividends = aapl_dividends[(aapl_dividends.index >= '2015-01-01') & (aapl_dividends.index <= '2023-12-31')]
msft_dividends = msft_dividends[(msft_dividends.index >= '2015-01-01') & (msft_dividends.index <= '2023-12-31')]

# Resample dividends data to quarterly frequency and sum up dividends for each quarter
aapl_quarterly_dividends = aapl_dividends.resample('Q').sum()
msft_quarterly_dividends = msft_dividends.resample('Q').sum()

# Plot dividends
plt.figure(figsize=(10, 6))
plt.plot(aapl_quarterly_dividends.index, aapl_quarterly_dividends, marker='o', linestyle='-', color='b', label='AAPL')
plt.plot(msft_quarterly_dividends.index, msft_quarterly_dividends, marker='o', linestyle='-', color='g', label='MSFT')
plt.title('Apple and Microsoft Quarterly Dividends (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Dividends (USD)')
plt.grid(True)
plt.xticks(aapl_quarterly_dividends.index, [date.strftime('%Y-%m') for date in aapl_quarterly_dividends.index], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd

# Define the ticker symbols
tickers = ['AAPL', 'MSFT']

# Fetch the data for both tickers
aapl = yf.Ticker(tickers[0])
msft = yf.Ticker(tickers[1])

# Fetch dividend data
aapl_dividends = aapl.dividends
msft_dividends = msft.dividends

# Filter the data for the period from 2015-01-01 to 2023-12-31
aapl_dividends = aapl_dividends[(aapl_dividends.index >= '2015-01-01') & (aapl_dividends.index <= '2023-12-31')]
msft_dividends = msft_dividends[(msft_dividends.index >= '2015-01-01') & (msft_dividends.index <= '2023-12-31')]

# Resample dividends data to yearly frequency and sum up dividends for each year
aapl_annual_dividends = aapl_dividends.resample('Y').sum()
msft_annual_dividends = msft_dividends.resample('Y').sum()

# Calculate the annual dividend growth rate
aapl_annual_growth_rate = aapl_annual_dividends.pct_change().dropna()
msft_annual_growth_rate = msft_annual_dividends.pct_change().dropna()

# Print the annual dividend growth rates
print("AAPL Annual Dividend Growth Rates:")
print(aapl_annual_growth_rate)

print("\nMSFT Annual Dividend Growth Rates:")
print(msft_annual_growth_rate)

# Convert to DataFrame for better readability
growth_rates_df = pd.DataFrame({
    'AAPL Annual Growth Rate': aapl_annual_growth_rate,
    'MSFT Annual Growth Rate': msft_annual_growth_rate
})

# Display the growth rates DataFrame
print("\nAnnual Dividend Growth Rates (2015-2023):")
print(growth_rates_df)

# Plot the annual dividend growth rates
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(aapl_annual_growth_rate.index, aapl_annual_growth_rate, marker='o', linestyle='-', color='b', label='AAPL')
plt.plot(msft_annual_growth_rate.index, msft_annual_growth_rate, marker='o', linestyle='-', color='g', label='MSFT')
plt.title('Annual Dividend Growth Rates (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Growth Rate')
plt.grid(True)
plt.xticks(aapl_annual_growth_rate.index, [date.strftime('%Y') for date in aapl_annual_growth_rate.index], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
