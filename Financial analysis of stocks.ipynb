```python
import pandas as pd
import matplotlib.pyplot as plt
from alpha_vantage.fundamentaldata import FundamentalData

# Define the ticker symbol and API key
ticker = 'BA'
api_key = 'YOUR_API_KEY'  # Replace with your API key from Alpha Vantage

# Create a FundamentalData object
fd = FundamentalData(key=api_key, output_format='pandas')

# Fetch the income statement
income_statement, _ = fd.get_income_statement_annual(ticker)

# Fetch the balance sheet
balance_sheet, _ = fd.get_balance_sheet_annual(ticker)

# Convert the index to datetime
income_statement['fiscalDateEnding'] = pd.to_datetime(income_statement['fiscalDateEnding'])
balance_sheet['fiscalDateEnding'] = pd.to_datetime(balance_sheet['fiscalDateEnding'])

income_statement.set_index('fiscalDateEnding', inplace=True)
balance_sheet.set_index('fiscalDateEnding', inplace=True)

# Convert relevant columns to numeric
income_statement['operatingIncome'] = pd.to_numeric(income_statement['operatingIncome'], errors='coerce')
income_statement['interestExpense'] = pd.to_numeric(income_statement['interestExpense'], errors='coerce')
balance_sheet['totalShareholderEquity'] = pd.to_numeric(balance_sheet['totalShareholderEquity'], errors='coerce')
balance_sheet['totalLiabilities'] = pd.to_numeric(balance_sheet['totalLiabilities'], errors='coerce')

# Filter the data for the period from 2015 to 2023
income_statement = income_statement[income_statement.index.year >= 2015]
balance_sheet = balance_sheet[balance_sheet.index.year >= 2015]

# Calculate Debt-to-Equity Ratio
debt_to_equity = balance_sheet['totalLiabilities'] / balance_sheet['totalShareholderEquity']

# Calculate Interest Coverage Ratio
interest_coverage = income_statement['operatingIncome'] / income_statement['interestExpense']

# Plot Debt-to-Equity and Interest Coverage Ratios with twin axes
fig, ax1 = plt.subplots(figsize=(8, 4))

color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Debt-to-Equity Ratio', color=color)
ax1.plot(debt_to_equity.index, debt_to_equity, marker='o', linestyle='-', color=color, label='Debt-to-Equity Ratio')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:red'
ax2.set_ylabel('Interest Coverage Ratio', color=color)  # we already handled the x-label with ax1
ax2.plot(interest_coverage.index, interest_coverage, marker='o', linestyle='-', color=color, label='Interest Coverage Ratio')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Boeing Debt-to-Equity and Interest Coverage Ratios (2015-2023)')
plt.xticks(debt_to_equity.index, debt_to_equity.index.year, rotation=45)
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the ticker symbol
ticker = 'BA'

# Fetch dividends data
ba = yf.Ticker(ticker)
dividends = ba.dividends

# Filter the data for the period from 2015 to 2023
dividends = dividends[(dividends.index >= '2015-01-01') & (dividends.index <= '2023-12-31')]

# Resample dividends data to yearly frequency and sum up dividends for each year
annual_dividends = dividends.resample('Y').sum()

# Plot dividends
plt.figure(figsize=(8, 4))
plt.plot(annual_dividends.index, annual_dividends, marker='o', linestyle='-', color='b', label='Annual Dividends')
plt.title('Boeing Annual Dividends (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Dividends')
plt.grid(True)
plt.xticks(annual_dividends.index, annual_dividends.index.year, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from alpha_vantage.fundamentaldata import FundamentalData

# Define the API key
api_key = 'YOUR_API_KEY'  # Replace with your API key from Alpha Vantage

# Function to fetch and process R&D expenses
def fetch_rd_expenses(ticker, api_key):
    fd = FundamentalData(key=api_key, output_format='pandas')
    income_statement, _ = fd.get_income_statement_annual(ticker)
    income_statement['fiscalDateEnding'] = pd.to_datetime(income_statement['fiscalDateEnding'])
    income_statement.set_index('fiscalDateEnding', inplace=True)
    correct_column_name = 'researchAndDevelopment'  # Column name for R&D expenses
    rd_expenses = pd.to_numeric(income_statement[correct_column_name], errors='coerce')
    rd_expenses = rd_expenses[(rd_expenses.index >= '2015-01-01') & (rd_expenses.index <= '2023-12-31')]
    rd_expenses = rd_expenses / 1e9  # Convert to billions
    return rd_expenses

# Fetch R&D expenses for each company
tickers = {'AMZN': 'Amazon', 'GOOG': 'Google', 'AAPL': 'Apple'}
rd_expenses_dict = {}

for ticker in tickers:
    rd_expenses_dict[ticker] = fetch_rd_expenses(ticker, api_key)

# Plot R&D expenses
plt.figure(figsize=(10, 6))

for ticker, company_name in tickers.items():
    plt.plot(rd_expenses_dict[ticker].index, rd_expenses_dict[ticker], marker='o', linestyle='-', label=f'{company_name} R&D Expenses')

plt.title('Annual R&D Expenses (2015-2023)')
plt.xlabel('Year')
plt.ylabel('R&D Expenses (in billions)')
plt.grid(True)
plt.xticks(rd_expenses_dict['AMZN'].index, rd_expenses_dict['AMZN'].index.year, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
