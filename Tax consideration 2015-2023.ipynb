```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with the actual tax rates
data = {
    'Year': list(range(2015, 2024)),
    'Dividend Basic': [10, 10, 7.5, 7.5, 7.5, 7.5, 7.5, 7.5, 8.75],
    'Dividend Higher': [32.5, 32.5, 32.5, 32.5, 32.5, 32.5, 32.5, 32.5, 33.75],
    'Dividend Additional': [37.5, 37.5, 38.1, 38.1, 38.1, 38.1, 38.1, 38.1, 39.35],
    'Capital Basic': [18, 18, 10, 10, 10, 10, 10, 10, 10],
    'Capital Higher': [28, 28, 20, 20, 20, 20, 20, 20, 20],
    'Capital Additional': [28, 28, 20, 20, 20, 20, 20, 20, 20]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Plot the data
plt.figure(figsize=(14, 8))

# Plot Dividend Tax Rates
plt.subplot(2, 1, 1)
plt.plot(df['Year'], df['Dividend Basic'], marker='o', label='Dividend Basic Rate')
plt.plot(df['Year'], df['Dividend Higher'], marker='o', label='Dividend Higher Rate')
plt.plot(df['Year'], df['Dividend Additional'], marker='o', label='Dividend Additional Rate')
plt.title('UK Dividend Tax Rates (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Tax Rate (%)')
plt.legend()
plt.grid(True)

# Plot Capital Gains Tax Rates
plt.subplot(2, 1, 2)
plt.plot(df['Year'], df['Capital Basic'], marker='o', label='Capital Gains Basic Rate')
plt.plot(df['Year'], df['Capital Higher'], marker='o', label='Capital Gains Higher Rate')
plt.plot(df['Year'], df['Capital Additional'], marker='o', label='Capital Gains Additional Rate')
plt.title('UK Capital Gains Tax Rates (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Tax Rate (%)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
